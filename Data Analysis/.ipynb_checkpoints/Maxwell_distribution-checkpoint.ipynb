```python
import pandas as pd
import numpy as np
import math
from scipy.optimize import curve_fit
from matplotlib import pyplot
```

```python
url = "velocities_0.csv"
data = pd.read_csv(url, header = None).T
bin_no = 15

pyplot.hist(data, bins = bin_no)

bars = pyplot.gca().patches

x = []
y = []

particles_no = 900

for bar in bars:
    x.append(bar.get_x())
    y.append(bar.get_height() / particles_no)

    
def max_bolt_distribution(x, a):
    return math.sqrt(2 / math.pi) * x**2 * np.exp(-x**2 / (2 * a**2)) / a**3

popt, _ = curve_fit(max_bolt_distribution, x, y)
a = popt
print('a = %.5f' % a)
pyplot.scatter(x, y)

x_fit = np.arange(min(x), max(x), 1)
y_fit = max_bolt_distribution(x_fit, a)

pyplot.plot(x_fit, y_fit, '--', color = 'red')
```
