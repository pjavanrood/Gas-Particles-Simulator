```python
import pandas as pd
import numpy as np
import math
from scipy.optimize import curve_fit
from matplotlib import pyplot
import random as rand
```

```python
url = "velocities_0.csv"
data = pd.read_csv(url, header = None)
data = data.T
bin_no = 15

pyplot.hist(data[0], bins = bin_no)

bars = pyplot.gca().patches

x = []
y = []

particles_no = 600

for bar in bars:
    x.append(bar.get_x())
    y.append(bar.get_height() / particles_no)

    
def max_bolt_distribution(x, a):
    return math.sqrt(2 / math.pi) * x**2 * np.exp(-x**2 / (2 * a**2)) / a**3

popt, _ = curve_fit(max_bolt_distribution, x, y)
a = popt
print('a = %.5f' % a)
pyplot.scatter(x, y)

x_fit = np.arange(min(x), max(x), 0.1)
y_fit = max_bolt_distribution(x_fit, a)
```

```python
fig, axs = pyplot.subplots(2)
len_df = len(data.columns)
for column in range(len_df):
    pyplot.hist(data[column], bins = bin_no)

    bars = pyplot.gca().patches

    x = []
    y = []
    for bar in bars:
        x.append(bar.get_x())
        y.append(bar.get_height() / particles_no)
    
    popt, _ = curve_fit(max_bolt_distribution, x, y)
    a = popt
    print('a = %.5f\n' % a)
    x_fit = np.arange(min(x), max(x), 0.1)
    y_fit = 600 * max_bolt_distribution(x_fit, a)
    axs[0].plot(x_fit, y_fit, '--', color = (column / len_df, rand.randrange(10) / 10, rand.randrange(10) / 10))

pyplot.show()
```
